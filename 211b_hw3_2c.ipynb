```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
import mpld3
mpld3.enable_notebook()
```

```python
def LHS(x):
    y = (x/2)*np.log((x+1)/(x-1))-1;
    return y

def RHS(x):
    f0 = 10.77;
    f1 = 6.25;
    y = (1+f1/3)/(f0*(1+f1/3)+x*x*f1);
    return y

lam = np.linspace(1.1,5,1000,endpoint=False)

plt.plot(lam,LHS(lam),'b-',lam,RHS(lam),'r-')
```
